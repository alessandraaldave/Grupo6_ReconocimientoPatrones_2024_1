## <center>Acerca de los datos </center>

​**Este conjunto de datos contiene 2126 registros de características extraídas de exámenes de Cardiotocograma, que luego fueron clasificados por un obstetra experto en 3 clases: "Normal", "Suspect" y "Pathological". El conjunto de datos contiene las siguientes características:**
    
* **baseline value:** Valor de línea base de la frecuencia cardíaca fetal (FHR) (latidos por minuto)   
* **accelerations:** Número de aceleraciones por segundo
* **fetal_movement:** Número de movimientos fetales por segundo    
* **uterine_contractions:** Número de contracciones uterinas por segundo
* **light_decelerations:** Número de desaceleraciones leves (LDs) por segundo
* **severe_decelerations:** Número de desaceleraciones severas (SDs) por segundo
* **prolongued_decelerations:** Número de desaceleraciones prolongadas (PDs) por segundo
* **abnormal_short_term_variability:** Porcentaje de tiempo con variabilidad anormal a corto plazo
* **mean_value_of_short_term_variability:** Valor medio de la variabilidad a corto plazo
* **percentage_of_time_with_abnormal_long_term_variability:** Porcentaje de tiempo con variabilidad anormal a largo plazo
* **mean_value_of_long_term_variability:** Valor medio de la variabilidad a largo plazo
* **histogram_width:** Ancho del histograma creado usando todos los valores de un registro
* **histogram_min:** Valor mínimo del histograma
* **histogram_max:** Valor máximo del histograma
* **histogram_number_of_peaks:** Número de picos en el histograma del examen
* **histogram_number_of_zeroes:** Número de ceros en el histograma del examen
* **histogram_mode:** Modo del histograma
* **histogram_mean:** Media del histograma
* **histogram_median:** Mediana del histograma
* **histogram_variance:** Varianza del histograma
* **histogram_tendency:** Tendencia del histograma
* **fetal_health:** Codificado como 1-Normal(Normal); 2-Sospechoso(Suspect); 3-Patológic(Pathological).

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix
from sklearn.model_selection import cross_val_score


```

```python
df = pd.read_csv("https://github.com/ulewis/Ejemplos/raw/main/Datos/fetal_health.csv")
df.head()  # vemos los 5 primeros datos del dataframe
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()

# Plot de la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de correlación')
plt.show()
```

```python
df.info()
```

```python
df['fetal_health'] = df['fetal_health'].map({1.0 :1, 2.0 :2, 3.0:3}).astype(int) #mapeo los targets
df.head()
```

```python
#Defino X, y
#Creo el clasificador
#Pongo data en.fit(X_train,y_train)
#Predict
```

```python
'''SET DE DATOS SIN SELECCIONAR CARACTERISTICAS'''
#X = df.iloc[:,:-1] #Selecciono features
#y = df['fetal_health'] #Selecciono target

'''ELIMINANDO UN FEATURE QUE NO CONTRIBUYEN A LA CLASIFICACIÓN'''
df = df.drop(columns=['histogram_mode','histogram_median','histogram_mean','mean_value_of_long_term_variability','accelerations'])
X = df.iloc[:,:-1] #Selecciono features
y = df['fetal_health'] #Selecciono target


#Hago uso de train_test_split para dividir dataset
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 101)
```

#Clasificacion con RF

```python
####   Random Forest del Challenge 6 (utilizando Grid Search CV)


from sklearn.model_selection import GridSearchCV
clasificador_RF_nuevo = RandomForestClassifier(random_state = 0)
parameters = {'min_samples_leaf':[5, 10, 20], 'max_depth':[5, 10], 'max_leaf_nodes': [10, 15, 20]}
grid_search_RF = GridSearchCV(estimator = clasificador_RF_nuevo, param_grid = parameters)
grid_search_RF.fit(X_train, y_train)


#Mejore estimaciones
mejor_RF_clasif = grid_search_RF.best_estimator_
RF_nuevo_prediccion =  mejor_RF_clasif.predict(X_test)
```

```python
print(mejor_RF_clasif)
```

```python
accuracy_R_nuevo = accuracy_score(y_test, RF_nuevo_prediccion)
precision_RF_nuevo = precision_score(y_test, RF_nuevo_prediccion, average='weighted')
recall_RF_nuevo = recall_score(y_test, RF_nuevo_prediccion, average='weighted')
F1_RF_nuevo = f1_score(y_test, RF_nuevo_prediccion, average='weighted')

print("Accuracy RF:", accuracy_R_nuevo)
print("Precision RF:", precision_RF_nuevo)
print("Recall RF:", recall_RF_nuevo)
print("F1-score RF:", F1_RF_nuevo)
```

```python
#### Random Forest del Challenge 4


clasificador_RF = RandomForestClassifier(max_depth=2, random_state=0)
clasificador_RF = clasificador_RF.fit(X_train, y_train)
RF_prediccion = clasificador_RF.predict(X_test)

accuracy_RF = accuracy_score(y_test, RF_prediccion)
precision_RF = precision_score(y_test, RF_prediccion, average='weighted')
recall_RF = recall_score(y_test, RF_prediccion, average='weighted')
F1_RF = f1_score(y_test, RF_prediccion, average='weighted')

print("Accuracy RF:", accuracy_RF)
print("Precision RF:", precision_RF)
print("Recall RF:", recall_RF)
print("F1-score RF:", F1_RF)

clasificador_CROSSVAL_RF = RandomForestClassifier(max_depth=2, random_state=0)
sccores_RF = sum(cross_val_score(clasificador_CROSSVAL_RF, X, y, cv=5))/5

print("Scores RF con CV = 5:", sccores_RF)

#Visualizacion
matriz_conf = confusion_matrix(y_test, RF_prediccion)

clases = ['Normal', 'Sospechoso', 'Patologico']
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_conf, annot=True, fmt="d", cmap="Blues", xticklabels=clases, yticklabels=clases)
plt.title("Matriz de Confusion")
plt.xlabel("Prediccion")
plt.ylabel("Real")
plt.show()
```

```python
### Implementacion de Pipeline para Random Forest sin Grid Search CV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipe = Pipeline([('scaler', StandardScaler()), ('svc', RandomForestClassifier(max_depth=2, random_state=0))])
pipe.fit(X_train, y_train).score(X_test, y_test)

```
