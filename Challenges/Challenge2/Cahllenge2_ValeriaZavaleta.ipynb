# **1. Introducción**

El cáncer de mama es una enfermedad que afecta a mujeres de todas las edades y es una de las principales causas de muerte relacionadas con el cáncer en el mundo. En el contexto peruano, el cáncer de mama representa un desafío significativo para la salud pública, con importantes implicaciones en términos de detección temprana, tratamiento y calidad de vida de las pacientes.

Según el Instituto Nacional de Enfermedades Neoplásicas (INEN) del Perú, el cáncer de mama es el tipo de cáncer más común en las mujeres peruanas, representando aproximadamente el 22.3% de todos los casos de cáncer en mujeres. Se estima que cada año se diagnostican alrededor de 6,000 nuevos casos de cáncer de mama en el país.

Además, las tasas de mortalidad por cáncer de mama en el Perú se han incrementado de manera preocupante. Según datos del Ministerio de Salud del Perú, el cáncer de mama es la segunda causa de muerte por cáncer en mujeres, después del cáncer de cuello uterino. Se estima que alrededor de 2,500 mujeres peruanas mueren cada año a causa de esta enfermedad.

Existen varios factores de riesgo asociados con el desarrollo de cáncer de mama en el Perú, incluyendo antecedentes familiares de la enfermedad, edad avanzada, obesidad, consumo de alcohol, falta de actividad física y exposición a ciertos factores ambientales y hormonales.

Además, las barreras en el acceso a servicios de salud, la falta de conciencia sobre la importancia de la detección temprana y el estigma asociado con el cáncer son desafíos adicionales que enfrentan las mujeres peruanas en la lucha contra esta enfermedad.


La detección temprana del cáncer de mama es fundamental para mejorar los resultados del tratamiento y reducir la mortalidad asociada con esta enfermedad. En el contexto peruano, donde las tasas de mortalidad por cáncer de mama siguen siendo preocupantes, los métodos de detección temprana, como la mamografía, el autoexamen de mama y la evaluación clínica regular, juegan un papel crucial en la identificación precoz de posibles anomalías y la intervención oportuna. La promoción de estos métodos de detección temprana, junto con el acceso equitativo a servicios de salud de calidad, es esencial para mejorar los resultados de salud de las mujeres peruanas y reducir la carga del cáncer de mama en el país.

# **2. Metodología**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importamos el archivo y visualizamos las primeras filas:
df_cancer = pd.read_csv("bc_data.csv")

# Para eliminar la primera y la útlima columna, pues su información es irrelevante:
df_cancer = df_cancer.drop(df_cancer.columns[0], axis=1)
df_cancer = df_cancer.drop(df_cancer.columns[-1], axis=1)

df_cancer.head()
```

### **2.3. Preprocesamiento**
Una vez que hemos determinado las variables redundantes y datos atípicos que no aportan al modelo, podemos proceder a suprimirlos.

En este caso, convendría retirar las siguientes columnas para evitar la redundancia: perimeter_mean, area_mean, radius_worst, perimeter_worst, area_worst, perimeter_se, area_se

```python
columnas_a_eliminar = ['perimeter_mean', 'area_mean', 'radius_worst', 'perimeter_worst', 'area_worst','perimeter_se', 'area_se']

# Eliminar las columnas con los nombres especificados
df_clean = df_cancer.drop(columnas_a_eliminar, axis=1)
df_clean.head(10)
```

```python
# Representación visual de cómo las variables se relacionan entre sí (matriz correlacional)
corr = df_clean.corr()
fig, ax = plt.subplots(figsize=(10, 10))
ax.matshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
sns.heatmap(corr,
    cmap=sns.diverging_palette(220, 10, as_cmap=True),
    vmin=-1.0, vmax=1.0,
    square=True, ax=ax)
plt.show()
```

```python
# Además limpiamos los datos atípicos
print('Data size before cleaning: ', len(df_cancer))
df_clean2 = df_clean[(df_clean['radius_mean'] < 22) & (df_clean['texture_mean'] < 32) & ((df_clean['smoothness_mean'] < 0.13) & (df_clean['smoothness_mean'] > 0.07)) &
                 (df_clean['compactness_mean'] < 0.21) & (df_clean['concavity_mean'] < 0.2) & (df_clean['concave points_mean'] < 0.16) & (df_clean['fractal_dimension_mean'] < 0.08) &
                 ((df_clean['symmetry_mean'] < 0.24) & (df_clean['symmetry_mean'] > 0.12)) & (df_clean['radius_se'] < 0.8) & (df_clean['texture_se'] < 2.2) & (df_clean['smoothness_se'] < 0.013) &
                 (df_clean['compactness_se'] < 0.06) & (df_clean['concavity_se'] < 0.08) & (df_clean['concave points_se'] < 0.028) & (df_clean['symmetry_se'] < 0.038) &
                 (df_clean['fractal_dimension_se'] < 0.008) & (df_clean['texture_worst'] < 43) & (df_clean['smoothness_worst'] < 0.190) & (df_clean['compactness_worst'] < 0.6) &
                 (df_clean['concavity_worst'] < 0.8) & (df_clean['symmetry_worst'] < 0.43) & (df_clean['smoothness_worst'] < 0.190) & (df_clean['fractal_dimension_worst'] < 0.125)]

print('Data size after cleaning: ', len(df_clean2))
```

```python
# Extraer la primera fila como etiquetas para facilitar el manejo de gráficos:
titles2 = df_clean2.columns.tolist()
titles2.pop(0) # Eliminamos el primer elemento, pues no evaluaremos el target

# Ploteo de diagramas de caja y bigotes, incluyendo los casos atípicos
plt.figure(figsize=(15,30))
count = 1
for feature in titles2:
    plt.subplot(10, 3, count)
    plt.boxplot(df_clean2[feature])
    plt.title(('Distribution of ' + feature))
    count += 1

plt.tight_layout()
plt.show()
```

# **3. Resultados**

### **3.1. Clasificación usando Guassian Naive Bayes**

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Suponiendo que tu DataFrame se llama df_clean2 y el target está en la primera columna
X = df_clean2.iloc[:, 1:]  # Características
y = df_clean2.iloc[:, 0]   # Target

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=142)

# Inicializar el clasificador Naive Bayes Gaussiano
gnb = GaussianNB()

# Entrenar el clasificador
gnb.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo: {:.2f}%".format(accuracy * 100))

# Obtener el informe de clasificación
report = classification_report(y_test, y_pred)

# Imprimir el informe de clasificación
print("Informe de clasificación:")
print(report)
```

### **3.2. Clasificación usando KNN Neighbours**

```python
from sklearn.neighbors import KNeighborsClassifier

# Inicializar el clasificador KNN
knn = KNeighborsClassifier(n_neighbors=5)  # Puedes ajustar el valor de n_neighbors según tus necesidades

# Entrenar el clasificador
knn.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = knn.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo: {:.2f}%".format(accuracy * 100))

# Obtener el informe de clasificación
report = classification_report(y_test, y_pred)

# Imprimir el informe de clasificación
print("Informe de clasificación:")
print(report)
```

# **4. Discusión**

* Precisión del modelo:

Ambos modelos muestran una precisión bastante alta, con KNN ligeramente por encima de Naive Bayes (96.20% vs. 93.67%). Esto indica que ambos son capaces de clasificar correctamente la mayoría de las instancias en el conjunto de prueba.
* Recall (sensibilidad):

Para la clase benigna (0), ambos modelos muestran una capacidad similar para identificar casos verdaderamente benignos (recall de aproximadamente 0.97).
Sin embargo, para la clase maligna (1), KNN supera significativamente a Naive Bayes (recall de 0.95 vs. 0.86). Esto significa que KNN es más efectivo para detectar casos malignos en comparación con Naive Bayes.
* F1-score:

El puntaje F1, que es una combinación de precisión y recall, es ligeramente más alto en KNN para ambas clases, lo que sugiere un mejor equilibrio entre precisión y recall en comparación con Naive Bayes.

 Los resultados muestran que KNN supera a Naive Bayes en términos de precisión global y capacidad para detectar casos malignos. Esto sugiere que KNN es más efectivo en la clasificación de cáncer de mama en este conjunto de datos específico.

La detección precisa de casos malignos es de suma importancia en el diagnóstico de cáncer de mama, ya que un diagnóstico erróneo podría tener graves consecuencias para el paciente.
